```python
key_val = {}
dij = 0
def blocker(filename):
    global l
    global f
    global test
    global dij
    l = []
    x1 = 7
    y1 = 7
    imgo = cv2.imread(filename)
    gray = cv2.cvtColor(imgo, cv2.COLOR_BGR2GRAY)
    linek = np.zeros((11, 11), dtype=np.uint8)
    linek[5, ...] = 1
    x = cv2.morphologyEx(gray, cv2.MORPH_OPEN, linek, iterations=1)
    gray -= x
    gray = cv2.bitwise_not(gray)
    # print("gray",gray)
    ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)
    while (True):
        rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (x1, y1))
        # print(rect_kernel)
        dilation = cv2.dilate(thresh1, rect_kernel, iterations=6)
        contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        # print("contours",contours)
        # print("hierarchy",hierarchy)
        test = len(contours)
        print("test",test)
        print(test)
        if (test < 10):
            x1 -= 1
            y1 -= 1
        elif test > 30:
            x1 += 1
            y1 += 1
        else:
            break
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        print("x",x)
        print("y",y)
        print("w",w)
        print("h",h)
        cv2.rectangle(imgo, (x, y), (x + w, y + h), (0, 255, 0), 2)
        # z = cv2.rectangle(imgo, (x, y), (x + w, y + h), (0, 255, 0), 2)
        roi = imgo[y:y+h, x:x+w]
        # v = cv2.imread(z)
        cv2.imwrite('./cropped_image/final_cropped' +str(dij)+ '.png',roi)
        dij += 1
x = blocker(img)
```

```python
import os
import cv2
l = []

# Specify the directory path containing the images
directory_path = '/home/teena/Documents/OCR_IMAGE_TEXT/cropped_image'

# Get the list of files in the directory
file_list = os.listdir(directory_path)

# Loop through each file in the directory
for file_name in file_list:
    # Construct the full path of the image file
    image_path = os.path.join(directory_path, file_name)

    # Read the image using OpenCV
    image = cv2.imread(image_path)
    st = pytesseract.image_to_string(image)
    cleaned_text = st.strip().replace('\n', ' ').replace('\x0c', '')
    print("cleaned_text",cleaned_text)
    l.append([cleaned_text])
print(l)
```

```python
def extract_information(data):
    bill_to = None
    ship_to = None
    terms_conditions = None
    unit_price = None
    description = None

    for item in data:
        text = item[0]

        if 'Bill' in text:
            bill_to = text.replace('Bill', '').strip()

        if 'Ship' in text:
            ship_to = text.replace('Ship To', '').strip()

        if 'Terms & Conditions' in text:
            terms_conditions = text.replace('Terms & Conditions', '').strip()

        if 'UNIT PRICE' in text:
            unit_price = text.replace('UNIT PRICE', '').strip()

        if 'DESCRIPTION' in text:
            description = text.replace('DESCRIPTION', '').strip()

    return bill_to, ship_to, terms_conditions, unit_price, description
```

```python
# Extract the information
bill_to, ship_to, terms_conditions, unit_price, description = extract_information(l)

# Print the results
print("Bill To:", bill_to)
print("Ship To:", ship_to)
print("Terms & Conditions:", terms_conditions)
print("UNIT PRICE:", unit_price)
print("DESCRIPTION:", description)
```
